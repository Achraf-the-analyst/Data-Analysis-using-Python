```python
df.isin(["?"]).sum()
```

```python
for x in df.index:
  if df.loc[x, "workclass"] == '?':
     df.loc[x, "workclass"] = "NaN"
```

```python
for x in df.index:
    if df.loc[x,"occupation"] == "?":
        df.loc[x,"occupation"] = "NaN"
```

```python
for x in df.index:
    if df.loc[x,"native-country"] == "?":
        df.loc[x,"native-country"] = "NaN"
```
